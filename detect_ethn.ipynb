```python
def preprocess_image(img, target_size=(224, 224)):
    img = load_image(img)
    base_img = img.copy()
    
    img = cv2.resize(img, target_size)
    img_pixels = image.img_to_array(img)
    img_pixels = np.expand_dims(img_pixels, axis = 0)
    img_pixels /= 255
    return img_pixels
```
